```python
#plotting in pandas
x=pd.Series(np.random.normal(size=50))
x.plot()
```

```python
x.cumsum().plot(grid=True)
```

```python
y=pd.DataFrame({'normal':np.random.normal(size=5),
                'gamma':np.random.gamma(1, size=5),
                'poisson':np.random.poisson(size=5)})
print(y)
y.cumsum(0).plot()
```

```python
y.cumsum(0).plot(subplots=True)
```

```python
y.cumsum(0).plot(secondary_y='normal')
```

```python
fig,axes=plt.subplots(nrows=1, ncols=3,figsize=(4,12))
for i,var in enumerate(['normal','gamma','poisson']):
         y[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
#saving figures to file
fig.savefig('figure.png')
```

```python
#simple line plots
fig=plt.figure()
ax=plt.axes()
s=np.linspace(0,20,100)
print(s)
ax.plot(s,np.sin(s));
```

```python
plt.plot(s,np.sin(s))
```

```python
#line colors
plt.plot(s, np.sin(s-0), color='red')
plt.plot(s, np.sin(s-1), color='pink')
plt.plot(s, np.sin(s-2), color='blue')
```

```python
#linestyle
plt.plot(s, np.sin(s-0), linestyle='-')
plt.plot(s, np.sin(s-1), linestyle=':')
plt.plot(s, np.sin(s-2), linestyle='-.')
```

```python
#adjusting the plots:axes limit
plt.plot(s, np.sin(s-0))
plt.xlim(-1,15)
plt.ylim(-2,2);
```

```python
#labeling plots
plt.plot(s, np.cos(s))
plt.title('cosine curve')
plt.xlabel('x')
plt.ylabel('cos(x)')
```

```python
x=np.linspace(1,10,50)
y=np.log(x)
plt.plot(x,y,'d',color='red')
```

```python
a=np.random.RandomState(7)
x=a.randn(100)
y=a.randn(100)
colors=a.rand(100)
sizes=1000*a.rand(100)
plt.scatter(x,y,c=colors, s=sizes, alpha=0.5,
            cmap='Reds')
plt.colorbar();
```

```python
#errorbars
x=np.linspace(0,10,100)
dy=0.1
y=np.sin(x)+dy*np.random.randn(100)
plt.errorbar(x,y,yerr=dy,fmt='.b');
```

```python

#histogram
plt.style.use("seaborn-white")
data=np.random.randn(10000)
```

```python
plt.hist(data)
```

```python
x=np.random.normal(0,0.8,1000)
y=np.random.normal(-2,1,1000)
kwargs=dict(histtypes='stepfilled', alpha=0.5, normed=True, bins=50)
plt.hist(x,**kwargs)
plt.hist(y, **kwargs)
```

```python
counts, bin_edges=np.histogram(data, bins=5)
print(counts)
```

```python
#two dimensional histograms
mean=[5,2]
cov=[[3,1],[2,0.5]]
x,y=np.random.multivariate_normal(mean, cov,1000).T
```

```python
plt.hist2d(x,y,bins=20,cmap='Reds')
cb=plt.colorbar()
```

```python
plt.hexbin(x,y,gridsize=20,cmap='Reds')
cb=plt.colorbar()
```
